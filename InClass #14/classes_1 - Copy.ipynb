## What is OOP?
OOP organizes code into **objects** that bundle together **data (attributes)** and **behavior (methods)**
- Ex1: a list object consists of the elements inside and the associated methods
- Ex2: Dog 🐶
  - Attributes: name, breed, age  
  - Methods: bark(), eat(), sleep()

OOP makes code easier to understand, reuse, and maintain.

## Defining a Class and Creating Objects
- A **class** defines a **blueprint** for creating objects. It defines the properties (attributes) and behaviors (methods) that the objects created from the class will have.
  -  A class is like a blueprint for a house, and objects are the actual houses built from that blueprint.
- An **object** is an instance of a class. For example, if Dog is a class, then my_dog = Dog() creates an object my_dog. 
- **Attributes** are Variables that belong to an object (e.g., name, age).
- `__init__` is a special method in Python classes, called the **constructor**, which is used to initialize objects when a class is instantiated
  - Inside `__init__`, the attributes are attached to the `self` object
- `self` refers to the specific object instance when defining a class
- **Methods:** Functions that belong to an object


```python
class Dog: # The first letter of classname is typically uppercase
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age

    def bark(self):
        print(f"{self.name} says woof!")

# Create instances
dog1 = Dog("Buddy", 3)
dog2 = Dog("Luna", 5)

dog1.bark()
dog2.bark()
```

```python
# Defining a class (blueprint)
class Cookie:
    def __init__(self, shape, flavor):
        self.shape = shape  # Attribute
        self.flavor = flavor  # Attribute

    def describe(self):
        print(f"This is a {self.shape}-shaped cookie with {self.flavor} flavor.")

# Creating objects (instances)
cookie1 = Cookie("star", "chocolate")
cookie2 = Cookie("circle", "vanilla")

# Calling a method
cookie1.describe()  # Output: This is a star-shaped cookie with chocolate flavor.
cookie2.describe()  # Output: This is a circle-shaped cookie with vanilla flavor.
```

```python
class Student:

  ## define the attributes
    def __init__(self, name = "Anonymous", major = "Engineering"):
        self.name = name
        self.major = major
        self.grades = []

  ## Have a method to add all the grades
    def add_grades(self, grades):
        self.grades = grades

  ## Have a method to add grades incrementally
    def add_grade(self, grade: list):
        self.grades.extend(grade) # pay attention to the difference between append() and extend

   ## Have a method to compute the average grades
    def compute_grades(self):
        avg = sum(self.grades)/len(self.grades)
        if avg >= 90:
            GPA = 'A'
        elif avg >= 80:
            GPA = 'B'
        elif avg >= 70:
            GPA = 'C'
        else:
            GPA = 'D'
        return avg, GPA



student = Student("Taylor", "music")
print(type(student))
print(student.name)
student.add_grade([100, 70, 90])
print(student.grades)
student.compute_grades()
print(student.compute_grades())
```

- Class vs Instance Attributes

```python
class Circle:
    pi = 3.1416  # class attribute (shared)

    def __init__(self, radius):
        self.radius = radius  # instance attribute (unique)

    def area(self):
        return Circle.pi * self.radius ** 2

c1 = Circle(2)
c2 = Circle(5)
print(c1.area(), c2.area())
```

**Key Difference:**
- *Instance attributes* → belong to individual objects.
- *Class attributes* → shared by all instances of a class.

## Dunder methods
- Special methods that start and end with two underscores, like: `__init__`, `__str__`, `__len__`, `__add__`
- They’re also called magic methods or Special methods, that let you customize how your class behaves with Python’s built-in operators, functions, and syntax.

| **Method**     | **When It’s Called**                            | **Example**                      |
| -------------- | ----------------------------------------------- | -------------------------------- |
| `__init__`     | When a new object is created                    | Initialize attributes            |
| `__str__`      | When `print()` or `str()` is used               | Return readable string           |
| `__repr__`     | When object is printed in shell / for debugging | Return developer-friendly string |
| `__len__`      | When `len(obj)` is called                       | Define object’s length           |
| `__add__`      | When using `+` operator                         | Define how two objects add       |
| `__sub__`      | When using `-` operator                         | Define subtraction               |
| `__eq__`       | When using `==`                                 | Define equality behavior         |
| `__lt__`       | When using `<`                                  | Define less-than comparison      |
| `__getitem__`  | When using `obj[index]`                         | Custom indexing                  |
| `__setitem__`  | When assigning `obj[index] = value`             | Custom item assignment           |
| `__call__`     | When object is called like a function           | Allow object to act callable     |
| `__del__`      | When object is deleted                          | Cleanup resources                |
| `__contains__` | When using `in`                                 | Define membership test           |




```python
class Student:
    def __init__(self, name, major):
        self.name = name
        self.major = major

    def __str__(self):
        return f"Student(name={self.name}, major={self.major})"

s = Student("Emma", 'CS')
print(s)
```

```python
class Item:

    def __init__(self, name: str, price: float, quantify = 0):
      # assign to self objects
        self.name = name
        self.price = price
        self.quantify = quantify

    def calculate_total_price(self):
        return self.price * self.quantify

    def calculate_sale_price(self, discount: float):
        return self.price * self.quantify * discount

    def __str__(self):
        return f"Item: (name={self.name}, price={self.price}, quantity = {self.quantify})"


item1 = Item("Phone",100)
print(type(item1))
print(item1.name)
print(item1)
#print(item1.name)
#print(item1.price)
#print(item1.calculate_total_price())
#print(item1.calculate_sale_price(0.7))
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"x coordinate is {self.x}, y coordinate is {self.y}"

p1 = Point(2, 3)
p2 = Point(4, 1)
print(p1 + p2)
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def __len__(self):
        return len(self.items)

cart = ShoppingCart()
cart.add_item("Milk")
cart.add_item("Eggs")
print(len(cart))  
```
