### Exercise 1 ‚Äî Student Class
Create a `Student` class with:
- Attributes: `name`, `major`, `gpa`
- Method: `honor_roll()` ‚Üí returns `True` if GPA ‚â• 3.5

Creat an student object with name Emma and GPA 3.7, print if Emma is on the Honor roll.

````python
class Student:
    def __init__(self, name, major, gpa):
        self.name = name
        self.major = major
        self.gpa = gpa

    def honor_roll(self):
        return self.gpa >= 3.5

emma = Student("Emma", "Engineering", 3.7)
print(emma.honor_roll())
````

### Exercise 2 ‚Äî Library System
Create a `Book` class with:
- Attributes: `title`, `author`, `available=True`
- Method: `borrow()` ‚Üí sets `available=False`
- Method: `return_book()` ‚Üí sets `available=True`

Test:
```python
book1 = Book("1984", "George Orwell")
book1.borrow()
print(book1.available)  # False
book1.return_book()
print(book1.available)  # True
```

````python
class Book:
    def __init__(self, title, author, available = True):
        self.title = title
        self.author = author
        self.available = available

    def borrow(self):
        self.available = False

    def return_book(self):
        self.available = True

book1 = Book("1984", "George Orwell")
book1.borrow()
print(book1.available)
book1.return_book()
print(book1.available)
````

### Exercise 3 ‚Äî Rectangle Class
Create a `Rectangle` class with:
- Attributes: width, height
- Methods:
    - `area()` ‚Üí returns area
    - `perimeter()` ‚Üí returns perimeter
    - `__str__()`‚Üí using formatted string to print the width and height

Then instantiate the Rectangle class using
```
rect = Rectangle(5, 3)
print(rect)
print(rect.area())       # 15
print(rect.perimeter())  # 16
```


````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def __str__(self):
        return f"Rectangle with width {self.width} and height {self.height}"

rect = Rectangle(5, 3)
print(rect)
print(rect.area())
print(rect.perimeter())
````

### Exercise 4 ‚Äî Pet Adoption Center
Create a `Pet` class with attributes:
- name, species, age, adopted (default False)
- Add methods:
  - `adopt()` ‚Üí marks the pet as adopted
  - `info()` ‚Üí prints a friendly message like:
   "Coco the cat is 2 years old and available for adoption." or "Coco the cat has been adopted."

Test the class using
```python
pet1 = Pet("Coco", "cat", 2)
pet1.info()
pet1.adopt()
pet1.info()
```



````python
class Pet:
    def __init__(self, name, species, age, adopted = False):
        self.name = name
        self.species = species
        self.age = age
        self.adopted = adopted

    def adopt(self):
        self.adopted = True
    
    def info(self):
        if not self.adopted:
            print(f"{self.name} the {self.species} is {self.age} years old and available for adoption.")
        else:
            print(f"{self.name} the {self.species} has been adopted.")

pet1 = Pet("Coco", "cat", 2)
pet1.info()
pet1.adopt()
pet1.info()
````
