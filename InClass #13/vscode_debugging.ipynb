## 1. What is VS Code?
**VS Code** is a cross-platform code editor made by Microsoft, which integrates several tools into one environment:
- 🧾 **Code editor** with syntax highlighting and auto-completion
- 🖥️ **Terminal** to run code inside the editor
- 🪲 **Debugger** to find and fix errors
- ⚙️ **Git integration** for version control
- 🧩 **Extensions** to add new features


**Why use VS Code for Python?**
- Lightweight and fast
- Excellent Python support through extensions
- Works great for large projects and debugging


### Step 1: Install VS Code
1. Go to [https://code.visualstudio.com/](https://code.visualstudio.com/)

### Example 1: Bubble Sort
- Repeatedly compare adjacent elements and swap them if they’re out of order.
- After each pass, the largest element “bubbles up” to the end.


For example, [5, 3, 8, 4]
- compare (5,3) → swap → [3,5,8,4]
- compare (5,8) → no swap
- compare (8,4) → swap → [3,5,4,8]
Next pass → [3,4,5,8] ✅

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):  # last i elements already sorted
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

nums = [5, 3, 8, 4, 2]
print("Bubble Sort:", bubble_sort(nums))
````

**Q:** What is the time complexity in terms of the length of the list?

Normally, Bubble Sort always makes n–1 passes, even if the list gets sorted early.
By adding a flag variable, we can detect if no swaps happened during a pass — which means the list is already sorted, and we can stop.

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        print(i)
        # Assume the list is already sorted
        sorted_flag = True

        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap the elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                sorted_flag = False  # Found unsorted pair
        
        # If no swaps happened, list is sorted
        if sorted_flag:
            print(f"Pass {i+1}: No swaps → already sorted ✅")
            break
        else:
            print(f"Pass {i+1}: Swaps happened → continue 🔁")

    return arr


nums = [2, 3, 4, 5, 6]
print("Original:", nums)
sorted_nums = bubble_sort(nums)
print("Sorted:", sorted_nums)
````

### Example 2: Binary Search
Find the index of a target element in a **sorted** list efficiently

````python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        print(f"Checking index {mid}, value {arr[mid]}")  # optional for tracing
        
        if arr[mid] == target:
            return mid   # found
        elif arr[mid] < target:
            low = mid + 1  # search right half
        else:
            high = mid - 1 # search left half
    return -1  # not found


arr = [1, 3, 4, 5, 7, 9, 11]
target = 7
result = binary_search(arr, target)

if result != -1:
    print(f"Found {target} at index {result}")
else:
    print("Not found")
````

## 4. Debugging in VS Code

- Open **Run and Debug** view, click **Run and Debug** to start
- Click left of a line number to toggle a **breakpoint**
- Step through code: Continue, Step Over, Step Into, Step Out
- Inspect variables, call stack, and watch expressions
- Use the Debug Console to evaluate code at a breakpoint


````python
def faulty_index(seq):
    # Bug: starts at 1, should start at 0
    total = 0
    for i in range(1, len(seq)):  # off-by-one bug
        total += seq[i]
    return total

print(faulty_index([1,2,3,4]))
````
