## Exercise 1 -- Encapsulation
Create a class `Chef` that hides a secret recipe ingredient.
1. The constructor accepts a `name` and initializes the private attribute `__secret_ingredient` to `"love ❤️"`.
2. `get_secret()` should print: `"Nice try, but the secret ingredient is classified!"` (do not return it).
3. `set_secret(new)` should only allow updates if `new` starts with `"magic"` (case-insensitive). Otherwise, print `"Access denied: Only magic ingredients allowed."`
4. `cook()` prints: `"Chef {name} is cooking with {magic mushrooms}!"`


```python
class Chef:
    def __init__(self, name):
        self.name = name
        self.__secret_ingredient = "love ❤️"
    
    def get_secret(self):
        print("Nice try, but the secret ingredient is classified!")
    
    def set_secret(self, new):
        if new.lower().startswith("magic"):
            self.__secret_ingredient = new
        else:
            print("Access denied: Only magic ingredients allowed.")
    
    def cook(self):
        print(f"Chef {self.name} is cooking with {self.__secret_ingredient}!")


chef = Chef("Emma")
chef.get_secret()
chef.set_secret("sugar")
chef.set_secret("magic mushrooms")
chef.cook()
```

## Exercise 2
Create a base class `Pet` and a subclass `Dog` that extends it.
- Base class `Pet(name)` with `speak()` that prints `"..."`.
- Subclass `Dog(name)`:
  - Calls `super().__init__(name)` in its constructor.
  - Overrides `speak()` to print `f"Woof! I am {name}!"`
  - Adds `fetch()` that prints `"I fetched the ball!"`

```python
class Pet:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("...")

class Dog(Pet):
    def __init__(self, name):
        super().__init__(name)
    
    def speak(self):
        print(f"Woof! I am {self.name}!")
    
    def fetch(self):
        print("I fetched the ball!")


d = Dog("Rex")
d.speak()
d.fetch()
```

## Exercise 3 — Multi-layer Inheritance: *The Royal Family*
Create a class hierarchy: `Person` → `Prince` → `King`.
- Person has an attribute `name `and method `introduce()` that prints `"I am just a person named <name>."`
- `Prince` inherits from `Person` and overrides `introduce()` to say:`"I am Prince <name>, heir to the throne."`
(Call `super().introduce() inside it.`)
- `King` inherits from `Prince` and overrides `introduce()` to say:
`"I am King <name>, ruler of the realm!"`
(Also call `super().introduce()`.)

```python
class Person:
    def __init__(self, name):
        self.name = name
    
    def introduce(self):
        print(f"I am just a person named {self.name}.")

class Prince(Person):
    def introduce(self):
        super().introduce()
        print(f"I am Prince {self.name}, heir to the throne.")

class King(Prince):
    def introduce(self):
        super().introduce()
        print(f"I am King {self.name}, ruler of the realm!")



arthur = King("Arthur")
arthur.introduce()
```

## Exercise 4 -- Composition
Create a `Robot` class that has an `Arm` object.
- Class `Arm` has methods:
  - `wave()` → prints `"Waving politely!"`
  - `grab()` → prints `"Grabbing coffee mug!"`
- Class `Robot` has:
  - An attribute `self.arm = Arm()`
  - A method `serve()` that calls `self.arm.wave()`, `self.arm.grab()`, then prints `"Here’s your coffee, master!"`

```python
class Arm:
    def wave(self):
        print("Waving politely!")
    
    def grab(self):
        print("Grabbing coffee mug!")

class Robot:
    def __init__(self):
        self.arm = Arm()
    
    def serve(self):
        self.arm.wave()
        self.arm.grab()
        print("Here’s your coffee, master!")

r = Robot()
r.serve()
```
